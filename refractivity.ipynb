```python
import glob
import datetime
import warnings

import numpy as np
import pandas as pd
import matplotlib.pyplot as pl
import matplotlib.dates as mdates

from scipy.signal import convolve
from scipy.signal import savgol_filter
from matplotlib.colors import LinearSegmentedColormap

warnings.simplefilter('ignore')
```

```python
def read_rs_compute_refra(infile):
    col_names = ['time', 'height', 'pressure', 'temperature', 'RH', 'dewpoint', 'direction', 'speed']    
    df = pd.read_csv(infile, skiprows=56, usecols=range(len(col_names)), names=col_names)
    
    dtime = datetime.datetime.strptime(infile[-19:-4], '%Y%m%d_%H%M%S')    
    z = np.linspace(0, 1e4, 101)

    h = df.height.values
    if len(h) == 0:
        return None
    
    p = df.pressure.values
    t = df.temperature.values
    RH = df.RH.values

    es = 6.1094 * np.exp(17.625 * t / (t + 243.04))
    ev = RH / 100 * es

    tkelvin = t + 273.15
    
    refractivity = 77.6 * p / tkelvin + 3.73e5 * ev / tkelvin ** 2
    refractivity = np.interp(z, h, refractivity)
    refractivity = savgol_filter(refractivity, 21, 3)
    
    dNdz = np.diff(refractivity)
    
    return dtime, z[:-1], dNdz / 0.1
```

```python
flist = sorted(glob.glob('/home/jupyter-vlouf/Voyages/current/voyage_specific/Radiosonde/Messages/VLMJ_YMC_RS41_isf_201912*.txt'))
```

```python
rslt = [read_rs_compute_refra(f) for f in flist]
rslt = [r for r in rslt if r is not None]

time, height, drefra = zip(*rslt)
refra = np.array(drefra)
z = height[0]
dtime = np.array(time, dtype='datetime64')
```

```python
refra_ls = np.array([-237, -158, -79, 0, 79])
colors = ['#F24738', '#F2CB05', '#05F29B', '#04D99D']
cm = LinearSegmentedColormap.from_list('nlist', colors, N=4)
```

```python
hours = mdates.HourLocator(interval=6)
days = mdates.DayLocator(interval=1)
```

```python
fig = pl.figure(figsize=(8, 4.5))

pl.pcolormesh(dtime, z, refra.T, cmap=cm, vmin=-237, vmax=79)
pl.title('Refractivity from RS.')
pl.ylabel('Altitude (m)')
pl.ylim(0, 2e3)
pl.axvline(dtime[2], color='b', linestyle='dotted', label='Sunset')
pl.axvline(dtime[6], color='k', linestyle='dashed', label='Sunrise')
[pl.axvline(x, color='b', linestyle='dotted', linewidth=1, alpha=0.75) for x in dtime[np.arange(11, len(dtime), 8)]]
[pl.axvline(x, color='k', linestyle='dashed', linewidth=1, alpha=0.75) for x in dtime[np.arange(15, len(dtime), 8)]]
pl.legend(loc=1)

cbar = pl.colorbar()
cbar.set_ticks([-198, -118, -37, 37])
cbar.set_ticklabels(['Ducting', 'Super-refraction', 'Normal-refraction', 'Sub-refraction'])

fig.tight_layout()
fig.autofmt_xdate()

pl.axes().xaxis.set_minor_locator(hours)
pl.axes().xaxis.set_major_locator(days)

pl.savefig('refractivity.png')
pl.show()
```
