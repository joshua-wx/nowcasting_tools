```python
from glob import glob
import subprocess

def gen_qsub(input_folder, output_folder, radar_type, walltime=10, py_script='',queue_mode=''):
    #build header
    conf_txt = \
    """#!/bin/bash
    #PBS -P kl02
    #PBS -q {qmode}
    #PBS -l walltime={time}:00:00
    #PBS -l mem={mem}GB
    #PBS -l wd
    #PBS -l ncpus=16
    #PBS -l jobfs=100GB
    #PBS -l storage=gdata/rq0+gdata/en0+gdata/kl02+gdata/en0+gdata/hj10
    #PBS -m abe
    export PATH="/g/data/en0/jss548/miniconda3:$PATH"
    source activate radar-prod
    python {python_script} -i {input_folder} -o {output_folder} -t {radar_type}
    """.format(time=walltime, mem=int(ncpu*2), input_folder=input_folder, output_folder=output_folder, radar_type=radar_type, python_script=py_script, qmode=queue_mode)
    return conf_txt

#main function
#paths

#CP2 S Band
vol_root = '/g/data/hj10/admin/cp2/level_1/s_band/sur'
gif_output_folder = '/g/data/hj10/admin/cp2/level_1/s_band/daily_gif'
radar_type = 'cp2'

#CP2 X Band
#vol_root = '/g/data/hj10/admin/cp2/level_1/x_band/sur'
#gif_output_folder  = '/g/data/hj10/admin/cp2/level_1/x_band/daily_gif'
#radar_type = 'cp2'

# #OPOL
# vol_root = '/g/data/hj10/admin/opol/level_1'
# gif_output_folder = '/g/data/hj10/admin/opol/level_1_daily_gif'
# radar_type = 'opol'

qsub_path    = "/home/548/jss548/dev/tmp/qsub_dailygif/"
py_script    = "/home/548/jss548/dev/projects/research_tools/daily_gif/daily_gif.py"
run_qsub_cmd = "qsub -k oe"
ncpu         = 16
walltime     = 10
run_flag     = True
qmode        = 'normal'
count        = 0

folder_list = sorted(glob(vol_root + '/*'))

#generate a date list
for i, vol_folder in enumerate(folder_list):
    #generate qsub
    f     = gen_qsub(vol_folder, gif_output_folder, radar_type, walltime, py_script, qmode)
    #write to file
    fname = '/'.join([qsub_path, "daily_gif_%s_%s.pbs" % (radar_type, str(i))])
    with open(fname, 'w') as fid:
        fid.write(f)
    #add to counter
    count += 1
    #run as required
    if run_flag == True:
        cmd           = ' '.join([run_qsub_cmd,fname])
        process       = subprocess.Popen(cmd.split(), stdout=subprocess.PIPE)
        output, error = process.communicate()
        print('runing ' + fname)
        print(output)

print('finished qsub generate for ',str(count), 'jobs')
```
